## Convert the height and weight datatypes to numerical forms 

### Let's start by the height column 

```python
fifa['Height'].unique()
```

We can see that there are value swith cm and with feet and inches

Let's create a function that will remove the 'cm' and do the right convertion for the feet values 

```python
def transform_height(df,col_name):
    
    # values with 'cm'
    df.loc[df[col_name].str.contains('cm'),col_name] = df.loc[df[col_name].str.contains('cm'),col_name].apply(lambda x: x.strip('cm'))
    
    # values with \'
    # cm = (feet*12+inches)*2.54
    series_feet_inches = df.loc[df[col_name].str.contains("\'"),col_name].str.strip('"').str.split('\'')
    
    for i in series_feet_inches.index : 
        df.loc[i,col_name] = round(((float(series_feet_inches[i][0]) * 12) + float(series_feet_inches[i][1]))*2.54)
        
    # convert the column type
    df[col_name] = fifa[col_name].astype(int)
    
    return df
```

```python
fifa = transform_height(fifa, 'Height')

# we can rename the column to specify that all the value is in kg
fifa = fifa.rename(columns={'Height' : 'Height(cm)'})

fifa['Height(cm)'].unique()
```

### Weight column

```python
 fifa['Weight'].unique()
```

We can see that there are values in kg and in lbs 

Let's define a function that will remove kg and lbs, and transform weight in lbs to kg

```python
def transform_weight(df,col_name): 
    
    # values in kg
    df.loc[df[col_name].str.contains('kg'),col_name] = df.loc[df[col_name].str.contains('kg'),col_name].apply(lambda x:x.strip('kg'))
    
    # values in lbs
    # kg = lbs * 0.45359237
    lbs_series = df.loc[df[col_name].str.contains('lbs'),col_name].str.strip('lbs')

    for i in lbs_series.index : 
        df.loc[i,col_name] = round(int(lbs_series[i]) * 0.45359237)
        
    # convert the column type 
    df[col_name] = df[col_name].astype(int)
    
    return df  
```

```python
fifa = transform_weight(fifa,'Weight')

# rename the column 
fifa = fifa.rename(columns={"Weight" : "Weight(kg)"})

fifa['Weight(kg)'].unique()
```

## Convert the 'Value', 'Wage' and "Release Clause' columns to numbers

Let's define a function that will transform each column

```python
def converting_column_to_int (df,col_name):
    """This function will change the Value,Wage and Release Clause colum type from 
    object to int, by removing the M, € and K characters on each row, and
    then cast the column type to int"""
    
    # values in million
    contain_M = df.loc[df[col_name].str.contains('M'),col_name].str.strip('M €')
    
    for i in contain_M.index : 
        df.loc[i,col_name] = float(contain_M[i])*1000000
        
    contain_K = fifa.loc[fifa[col_name].str.contains('K')==True,col_name].str.strip('K €')
    
    # values with K 
    for i in contain_K.index : 
        df.loc[i,col_name] = float(contain_K[i])*1000
        
    # 0€ value
    df.loc[df[col_name].str.contains('€')==True,col_name] = df.loc[df[col_name].str.contains('€')==True,col_name].apply(lambda x: x.strip('€'))
    
    # convert the column type 
    df[col_name] = df[col_name].astype(int)
    return df
```

```python
fifa = converting_column_to_int(fifa,'Value')
fifa = converting_column_to_int(fifa,'Wage')
fifa = converting_column_to_int(fifa,'Release Clause')
```

## Convert the Hits column to integer 

```python
fifa['Hits'].unique()
```

We can see that there are values with 'K' in the column , and also nan value

Let's transform those values before coverting the column type 

```python
def convert_hits(df,col_name): 
    """ This function will remove the 'K' from each row in the column 
    and fill nan value with '0' so that we can easily convert the column type to int"""
    
    contain_k = fifa.loc[fifa[col_name].str.contains('K')==True,col_name].str.strip('K')
    
    for i in contain_k.index : 
        df.loc[i,col_name] = float(contain_k[i]) * 1000
        
    # filling the nan values
    df['Hits'] = df['Hits'].fillna('0')
    
    df[col_name] = df[col_name].astype(int)
    
    return df
```

```python
fifa = convert_hits(fifa,'Hits')
fifa['Hits'].dtypes
```

## Some columns have 'star' characters. Strip those columns of these stars and make the columns numerical

### Those columns are W/F, SM, IR

```python
fifa['W/F'].unique(), fifa['SM'].unique(), fifa['IR'].unique()
```

We can see that the 'W/F' and 'IR' columns have the same values, the numbers in each column are seperated from the star by a space

Let's define a function that will remove the stars in the W/F and IR columns

```python
def strip_stars(df,col_name) : 
    """ This function will split all the value of the col_name on the ' 'space 
    and then CAST the type of the column type to integer"""
    df[col_name] = df[col_name].apply(lambda x :x.split(' ')[0])
        
    #changing the column type
    df[col_name] = df[col_name].astype(int)
    
    return df
```

```python
fifa = strip_stars(fifa,'W/F')
fifa = strip_stars(fifa,'IR')
```

```python
fifa['W/F'].dtypes, fifa['IR'].dtypes
```

Let's transform the last column 'SM'

```python
fifa['SM'] = fifa['SM'].apply(lambda x : x[0])
fifa['SM'] = fifa['SM'].astype(int)
fifa['SM'].dtypes
```
